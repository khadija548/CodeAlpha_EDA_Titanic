## 1. Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 2. Load Dataset

```python
df = pd.read_csv('/kaggle/input/titanic/train.csv')
df.head()
```

## 3. Dataset Overview

```python
df.info()
```

```python
df.describe()
```

## 4. Handling Missing Values (Age)

```python
df.isnull().sum()
```

```python
df["Age"] = df["Age"].fillna(df["Age"].mean())
df.isnull().sum()
```

## 5. Exploratory Data Analysis (Visualizations)

### Survival Count

```python
sns.countplot(x="Survived", data=df)
plt.title("Survival Count")
plt.show()
```

### Survival Count Analysis
### Insight:
More passengers did not survive than survived.

### Gender vs Survival

```python
sns.countplot(x="Sex", hue="Survived", data=df)
plt.title("Gender vs Survival")
plt.show()
```

### Insight:
Survival rates differ significantly by gender. Female passengers show a noticeably higher survival rate compared to male passengers, suggesting that gender played an important role in survival outcomes.


### Age Distribution Analysis

```python
sns.histplot(df["Age"], bins=30)
plt.title("Age Distribution")
plt.show()
```

### Insight:
The age distribution indicates that most passengers were young adults, with fewer elderly passengers. The data appears slightly right-skewed, showing a concentration of individuals in lower age ranges.


### Correlation Analysis (Heatmap)

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

### Insight:
The correlation heatmap highlights relationships between numerical features. Passenger class (Pclass) shows a noticeable relationship with survival, suggesting that socioeconomic factors may have influenced survival probability.


## Key Findings

1. The majority of passengers did not survive the disaster.
2. Female passengers had a significantly higher survival rate than males.
3. Most passengers were young adults based on the age distribution.
4. Passenger class shows a meaningful relationship with survival outcomes.
5. Missing values in the Age feature were handled using mean imputation to improve data quality.
